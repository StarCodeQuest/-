```python
# Identifying numerical columns in the dataset
numerical_cols = X.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Define a function to detect outliers using 3-sigma rule
def detect_outliers_3sigma(series):
    mean = series.mean()
    std = series.std()
    outliers_mask = (series < mean - 3 * std) | (series > mean + 3 * std)
    return outliers_mask, mean, std

# Apply the function to each numerical column and collect results
outliers_info = {}
for col in numerical_cols:
    outliers_mask, mean, std = detect_outliers_3sigma(data[col])
    outliers_info[col] = {
        'mean': mean,
        'std': std,
        'outliers_count': outliers_mask.sum(),
        'outliers_index': data[outliers_mask].index.tolist()
    }

# Convert the results into a DataFrame for better presentation
outliers_summary = pd.DataFrame.from_dict(outliers_info, orient='index', columns=['mean', 'std', 'outliers_count', 'outliers_index'])
outliers_summary
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# 生成一个遮罩来覆盖上三角形部分，因为矩阵是对称的

# 加载数据
# data = pd.read_csv('您的数据文件路径')

# 计算斯皮尔曼相关系数矩阵
spearman_corr = data.corr(method='spearman')
mask = np.triu(np.ones_like(spearman_corr, dtype=bool))
plt.figure(figsize=(15,15))
sns.heatmap(spearman_corr,mask=mask,  vmax=1, square=True)

plt.title('各个变量之间的相关性')
```

```python
import pandas as pd
import numpy as np


# 初始化一个列表来存储结果
high_corr_list = []

# 遍历上三角矩阵来找到相关系数大于0.9的变量对（不包括对角线）
for i, j in zip(*np.triu_indices_from(spearman_corr, k=1)):
    if spearman_corr.iloc[i, j] > 0.9:
        # 保存相关系数（保留5位小数），变量1，变量2
        high_corr_list.append([round(spearman_corr.iloc[i, j], 5), data.columns[i], data.columns[j]])

# 将结果转换为DataFrame
high_corr_df = pd.DataFrame(high_corr_list, columns=['相关性系数', '变量1', '变量2'])

high_corr_df


```
